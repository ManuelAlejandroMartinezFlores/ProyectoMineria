```python
import pandas as pd
```

```python
df = pd.read_csv("gtm-ndvi-adm2-full.csv")
df.shape
```

```python
df.head()
```

```python
df.n_pixels.unique()
```

```python
df["mes"] = df.date.map(lambda x: int(x.split("-")[1]))
df["anio"] = df.date.map(lambda x: int(x.split("-")[0]))
df["dia"] = df.date.map(lambda x: int(x.split("-")[2]))
df.head()
```

```python
c = pd.read_csv("guatemala_lat_lon.csv")
c.head()
```

```python
c = c[["ADM2_PCODE", "ADM1_ES", "ADM2_ES", "x", "y"]]
c.rename(columns={"ADM1_ES": "departamento", "ADM2_ES": "municipio", "x":"lon", "y": "lat"}, inplace=True)
c.head()
```

```python
merged = df.merge(c, "inner", on="ADM2_PCODE")
merged.head()
```

```python
merged.describe()
```

```python
df.shape, merged.shape
```

```python
merged.to_csv("ndvi-completo.csv", index=False)
```
