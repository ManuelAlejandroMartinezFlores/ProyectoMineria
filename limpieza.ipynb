```python
import pandas as pd
```

```python
df = pd.read_csv("precios_terminal.csv")
df.head()
```

```python
df.unit.unique()
```

```python
df["weight"] = df.unit.map(lambda x: float(x.split()[0]) if "KG" in x else float(x.split()[0])/2.2)
df.head()
```

```python
df["year"] = df.date.map(lambda x: int(x.split("-")[0]))
df["month"] = df.date.map(lambda x: int(x.split("-")[1]))
df.head()
```

```python
df["exchange_rate"] = df.gtqprice / df.usdprice
df.head()
```

```python
df["gtqprice_per_kg"] = df.gtqprice / df.weight
df["usdprice_per_kg"] = df.usdprice / df.weight
df.head()
```
